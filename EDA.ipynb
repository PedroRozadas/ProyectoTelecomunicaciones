```python
df_predic_internet_2028.rename(columns={'Internet users in Argentina 2019-2028' : 'Año',
    'Unnamed: 1' : 'Número de usuarios de internet en Argentina'}, inplace=True)
```

```python
df_predic_internet_2028.drop(index=[0,1], inplace=True)
```

```python
df_predic_internet_2028.sort_values(by='Año', ascending=False, inplace=True)
```

```python
df_predic_internet_2028
```

Calculo la variación porcentual de usuarios en internet, por año.

```python
x = df_predic_internet_2028['Número de usuarios de internet en Argentina']
df_predic_internet_2028['Variación porcentual'] = round((x - x.shift(-1)) / x.shift(-1) * 100, 2)
```

```python
a = df_predic_internet_2028.sort_values(by='Año', ascending=True)
df_predic_internet_2028['Suma variación porcentual'] = round(a['Variación porcentual'].cumsum(), 2)
```

Elimino duplicados si los hay.

```python
df_internet_penetracion.groupby('Año')['Accesos por cada 100 hogares'].describe()
```

```python
df_2022_3 = df_internet_penetracion.loc[(df_internet_penetracion['Año'] == 2022) & (df_internet_penetracion['Trimestre'] == 3)]
df_2022_3['Accesos por cada 100 hogares'].plot.box(grid=True)
plt.title('Accesos por cada 100 hogares, por provincia, trimestre 3, año 2022.')
plt.show()
```

```python
santa_cruz.describe()
```

Agrupo los datos por provincia y obtengo los valores de la columna "Accesos por cada 100 hogares" de un trimestre anterior.

```python
tr_anterior = df_internet_penetracion.groupby(['Provincia'])['Accesos por cada 100 hogares'].shift(-1)
```

Calculo el aumento del 2% del mismo.

```python
aumento_2 = tr_anterior * 1.02
```

Genero una nueva columna con la tasa de penetración esperada del 2% redondeada.

```python
df_internet_penetracion.loc[:, 'Tasa de penetración esperada'] = round(aumento_2, 2)
```

En el caso del último trimestre, normalizo llenando los NaN con el valor de 'Accesos por cada 100 hogares' del mismo.

```python
df_internet_penetracion.loc[:, 'Tasa de penetración esperada'] = df_internet_penetracion.loc[:, 'Tasa de penetración esperada'].fillna(df_internet_penetracion['Accesos por cada 100 hogares'])
```

Obtengo la diferencia entre el indicador de aumento del 2% con el valor de Accesos por cada 100 hogares del trimestre en cuestión.

```python
df_internet_penetracion['Desviación de penetración'] = round(df_internet_penetracion['Tasa de penetración esperada'] - df_internet_penetracion['Accesos por cada 100 hogares'], 2)
```

```python
df_internet_penetracion.loc[df_internet_penetracion['Provincia'] == 'Buenos Aires'].head()
```

```python
df_internet_penetracion.to_csv('df_internet_penetracion.csv', index= False)
```

Calculo la variación porcentual, por trimestre, por provincia.

```python
variacion_pct = df_internet_penetracion.groupby(['Provincia'])['Accesos por cada 100 hogares'].apply(lambda x: (x - x.shift(-1)) / x.shift(-1) * 100).rename("Variación porcentual")
```

```python
variacion_pct = variacion_pct.reset_index()
```

```python
variacion_pct.set_index('level_1', inplace=True)
```

```python
variacion_pct.drop(columns='Provincia', inplace=True)
```

```python
variacion_pct = variacion_pct.sort_index()
```

```python
variacion_pct.head()
```

```python
df_internet_penetracion = df_internet_penetracion.merge(variacion_pct, how='inner', left_index=True, right_index=True)
```

Promedio de porcentaje de aumento en el último año, cada 100 hogares, por provincia.

Agrupo filas por provincia, me quedo con el último año.

```python
df_internet_provincia_2022 = df_internet_penetracion.groupby(['Provincia']).apply(lambda x: x[x['Año'] == 2022])
```

Uso rolling con ventana de tamaño 3 tomando el promedio.

```python
df_internet_provincia_2022['Promedio variación porcentual'] = df_internet_provincia_2022['Variación porcentual'].rolling(3).mean()
```

```python
display(df_internet_provincia_2022)
```

Creo un datasets con el Promedio variación porcentual de cada provincia en el último año.

```python
col_index = df_internet_provincia_2022.columns.get_loc("Promedio variación porcentual")
df_prom_var_pct_2022 = df_internet_provincia_2022.iloc[2::3, col_index]
```

```python
df_prom_var_pct_2022 = df_prom_var_pct_2022.reset_index()
df_prom_var_pct_2022.drop(columns='level_1', inplace=True)
```

Promedio de porcentaje de aumento en los últimos dos años, cada 100 hogares, por provincia.

```python
df_internet_provincia_2021_2022 = df_internet_penetracion.groupby(['Provincia']).apply(lambda x: x[(x['Año'] == 2021) | (x['Año'] == 2022)])
```

```python
df_internet_provincia_2021_2022['Promedio variación porcentual'] = df_internet_provincia_2021_2022['Variación porcentual'].rolling(7).mean()
```
